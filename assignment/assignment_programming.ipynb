```python
#1. Import Numpy
import numpy as np
import matplotlib.pyplot as plt

#2. Grid
x = np.linspace(0,1,50)

#3. Numpy arrays y and z
y = np.log(x)
z = np.exp(x)

#4. Two scatter plots
plt.scatter(x,y)
plt.scatter(x,z)

#5. Showing the plot
plt.show()
```

```python
#6. Repeat steps 1-4
x = np.linspace(0,1,50)
y = np.log(x)
z = np.exp(x)

#7. Adding labels
plt.scatter(x,y, label ='natural log')
plt.scatter(x,z, label='exponential')
plt.xlabel("X")
plt.ylabel("Y")
plt.title('natural log and exponential functions')

#7. Adding a legend
plt.legend(loc='lower right')
plt.show()
```

```python
#8. New grid
x = np.linspace(-6.5,6.5,100)

#9. Sine and cosine functions
y = np.sin(x)
z = np.cos(x)

#10. Plotting values for the function with the values on the grid
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()
```

```python
#11. Making this a line graph instead
x = np.linspace(-6.5,6.5,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, label ='sin')
plt.plot(x,z, label='cos')
plt.show()
```

```python
#12. Final plot
x = np.linspace(-6.5,6.5,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, label ='sin')
plt.plot(x,z, label='cos')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc='lower left')
plt.title('sin and cos')
plt.show()
```
